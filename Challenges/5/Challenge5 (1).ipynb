| **Columna**                   | **Tipo**  | **Significado**                                                                                   |
|------------------------------|-----------|----------------------------------------------------------------------------------------------------|
| `Molecular_Weight`           | Numérico  | Peso molecular total en Daltons (g/mol), incluye todos los átomos.                                |
| `Polar_Area`                 | Numérico  | Área polar superficial (TPSA), relacionada con solubilidad y permeabilidad.                       |
| `Complexity`                 | Numérico  | Medida de la complejidad estructural de la molécula (ramificación, ciclos, estereocentros, etc.). |
| `XLogP`                      | Numérico  | Logaritmo del coeficiente de partición octanol/agua. Mide lipofilicidad.                          |
| `Heavy_Atom_Count`           | Numérico  | Número de átomos pesados (no H).                                                                  |
| `H-Bond_Donor_Count`         | Numérico  | Número de átomos capaces de donar un enlace de hidrógeno (como OH o NH).                          |
| `H-Bond_Acceptor_Count`      | Numérico  | Número de átomos capaces de aceptar un enlace de hidrógeno (como O, N con pares libres).          |
| `Rotatable_Bond_Count`       | Numérico  | Número de enlaces simples no rígidos. Relacionado con la flexibilidad molecular.                  |
| `Exact_Mass`                 | Numérico  | Masa exacta calculada con isótopos más abundantes (alta precisión).                               |
| `Monoisotopic_Mass`          | Numérico  | Igual que `Exact_Mass`, suele usarse en espectrometría de masas.                                  |
| `Charge`                     | Numérico  | Carga neta formal de la molécula.                                                                 |
| `Covalent_Unit_Count`        | Numérico  | Número de fragmentos moleculares conectados covalentemente (por ej., sales).                      |
| `Isotopic_Atom_Count`        | Numérico  | Número de átomos con isótopos específicos (no comunes).                                           |
| `Total_Atom_Stereo_Count`    | Numérico  | Total de centros estereogénicos en átomos (quiralidad).                                           |
| `Defined_Atom_Stereo_Count`  | Numérico  | Cuántos estereocentros están definidos.                                                           |
| `Undefined_Atom_Stereo_Count`| Numérico  | Cuántos estereocentros están indefinidos.                                                         |
| `Total_Bond_Stereo_Count`    | Numérico  | Número de estereoisomerismos tipo E/Z (dobles enlaces).                                           |
| `Defined_Bond_Stereo_Count`  | Numérico  | Cuántos están definidos como E o Z.                                                               |
| `Undefined_Bond_Stereo_Count`| Numérico  | Cuántos no tienen definida la geometría (E/Z).                                                    |

De la tabla vamos a seleccionar tres **descriptores moleculares** particularmente importantes para una interpretación futura con aplicaciones tecnológicas, científicas o farmacológicas:

1. **`Polar_Area` (Área Polar Superficial)**: Es crucial porque influye directamente en la **permeabilidad celular** y la **absorción oral** de una molécula, lo cual es vital para evaluar su viabilidad como fármaco o compuesto bioactivo.

2. **`XLogP`**: Mide la **lipofilicidad** de una molécula, un factor determinante en su capacidad para atravesar membranas biológicas. Es clave en el diseño de fármacos, donde el balance entre solubilidad en agua y en lípidos puede definir la eficacia de un compuesto.

3. **`H-Bond_Donor_Count` y `H-Bond_Acceptor_Count`**: (cuentan como uno por su relación directa) son esenciales porque determinan la **capacidad de interacción mediante enlaces de hidrógeno**, algo fundamental para la **afinidad y especificidad de unión a blancos moleculares** como proteínas o ADN.

```python
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(df_new)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

pca = PCA()
pca.fit(X_scaled)

explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(10, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance.cumsum(), marker='o')
plt.title('Varianza acumulada explicada por PCA')
plt.xlabel('Número de componentes')
plt.ylabel('Varianza acumulada')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Análisis de la Gráfica de Varianza Acumulada - PCA

La gráfica presentada muestra la varianza acumulada explicada por el Análisis de Componentes Principales (PCA) en función del número de componentes seleccionados. Este análisis nos permite identificar cuántas dimensiones son necesarias para conservar la mayor parte de la información del conjunto de datos original.

### Interpretación de la Gráfica

Cada punto en el eje horizontal representa una componente principal (PC) añadida al modelo, mientras que el eje vertical muestra el porcentaje acumulado de la varianza explicada. A medida que se incrementa el número de componentes, la varianza explicada se acerca al 100%, aunque con rendimientos decrecientes.

La forma ascendente de la curva indica que cada nueva componente aporta menos información que la anterior. La gráfica permite observar un punto de inflexión a partir del cual las nuevas componentes ya no contribuyen significativamente al total de varianza explicada.

### Dimensiones necesarias para conservar el 95% de la varianza

Al observar la curva, se identifica que la varianza acumulada alcanza el 95% aproximadamente con las primeras 13 componentes principales. Por tanto, concluimos que:

**Nuestro grupo requiere 13 componentes principales para conservar el 95% de la varianza del conjunto de datos original.**

Este valor representa un buen equilibrio entre reducción de dimensionalidad y retención de la información relevante.

### Relevancia del resultado

La reducción de dimensionalidad mediante PCA es fundamental para:

- Simplificar el análisis de datos de alta dimensión.
- Reducir el riesgo de sobreajuste en modelos predictivos.
- Mejorar la eficiencia computacional.
- Facilitar la visualización de los datos.

En este caso, al conservar solo 13 componentes, se logra mantener la mayor parte de la información del conjunto de datos original, permitiendo futuros análisis más eficientes sin pérdida significativa de contenido informativo.


## ¿Cómo usaríamos KMeans para reducir la dimensionalidad?

En nuestro grupo, utilizamos una estrategia combinada para reducir la dimensionalidad del conjunto de datos de forma más efectiva. En primer lugar, aplicamos **PCA (Análisis de Componentes Principales)** para reducir la cantidad de variables, conservando únicamente las 10 componentes principales que explican la mayor parte de la varianza. Esta etapa nos ayudó a eliminar información redundante y a simplificar los datos, manteniendo su estructura esencial.

Una vez reducido el número de dimensiones con PCA, aplicamos el algoritmo **KMeans** para agrupar los datos en clusters. Aunque KMeans no reduce directamente la dimensionalidad, cumple un rol importante al identificar patrones o grupos en el espacio reducido, permitiendo analizar la estructura del dataset con menor complejidad.

De esta forma, la combinación PCA + KMeans no solo mejora la eficiencia del clustering, sino que también hace que los resultados sean más interpretables y visualmente representables. En resumen, KMeans complementa el proceso de reducción de dimensionalidad al facilitar la agrupación de datos ya simplificados.


```python
components = pca.components_
feature_names = df_new.columns

pc_df = pd.DataFrame({
    'Feature': feature_names,
    'PC1': components[0],
    'PC2': components[1]
})

pc_df['PC1_abs'] = pc_df['PC1'].abs()
pc_df['PC2_abs'] = pc_df['PC2'].abs()

print("Top 10 variables más influyentes en PC1:")
print(pc_df.sort_values(by='PC1_abs', ascending=False)[['Feature', 'PC1']].head(10))

print("\nTop 10 variables más influyentes en PC2:")
print(pc_df.sort_values(by='PC2_abs', ascending=False)[['Feature', 'PC2']].head(10))
```

# K-means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples

k_range = range(2, 10)
inertias = []
silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=13)
    labels = kmeans.fit_predict(X_pca[:, :10]) 
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_pca[:, :10], labels))

fig, axs = plt.subplots(1, 2, figsize=(14, 5))

axs[0].plot(k_range, inertias, 'bo-')
axs[0].set_title("Codo")
axs[0].set_xlabel("Número de Clusters (k)")
axs[0].set_ylabel("Inercia")
axs[0].grid(True)
axs[1].plot(k_range, silhouette_scores, 'go-')
axs[1].set_title("Silueta")
axs[1].set_xlabel("Número de Clusters (k)")
axs[1].set_ylabel("Silhouette Score")
axs[1].grid(True)
plt.tight_layout()
plt.show()
```

Segun el plot del metodo del codo, usaremos k=3 como punto de inflexion.

```python
X_pca_10 = X_pca[:, :10]
kmeans = KMeans(n_clusters=3, random_state=13)
labels = kmeans.fit_predict(X_pca_10)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='Set2', s=10)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clusters con KMeans sobre PCA")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.cm as cm

score = silhouette_score(X_pca_10, labels)
print(f"Silhouette Score para KMeans (k=3): {score:.4f}")
```

# DBSCAN

```python
from sklearn.neighbors import NearestNeighbors
min_samples = 10
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors_fit = neighbors.fit(X_pca_10)  
distances, indices = neighbors_fit.kneighbors(X_pca_10)

k_distances = np.sort(distances[:, -1])

plt.figure(figsize=(8, 5))
plt.plot(k_distances)
plt.title(f"Gráfico de k-distancias (k = {min_samples})")
plt.xlabel("Instancias")
plt.ylabel(f"Distancia al {min_samples}° vecino más cercano")
plt.grid(True)
plt.tight_layout()
plt.show()
```

El codo está aproximadamente entre 3 y 5 unidades en el eje y.

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=5, min_samples=150)
db_labels = dbscan.fit_predict(X_pca_10)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=db_labels, cmap='tab10', s=10)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clusters con DBSCAN sobre PCA")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
score = silhouette_score(X_pca[:, :10], db_labels, sample_size=1000, random_state=13)
print(f"Silhouette Score para DBSCAN: {score:.4f}")
```

DBSCAN logró un buen agrupamiento (silhouette score ≈ 0.72). Esto indica que los grupos están bien formados y separados.
